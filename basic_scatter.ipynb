```python
plotly.offline.init_notebook_mode(connected=True)
```

```python
np.random.seed(42)
random_x = np.random.randint(1,101,100)
random_y = np.random.randint(1,30,100)

data = [go.Scatter(x = random_x, 
                   y = random_y, 
                   mode ='markers',
                   marker=dict(size = 12, 
                               color ='rgb(144,43,144)', 
                               symbol='pentagon'),
                               line = {'width' : 2}
                  )]
layout = go.Layout(title='Random aweseomnes', 
                    xaxis={'title':'My X Axis'},
                    yaxis = dict(title='My Y Axis'),
                    hovermode = 'closest')
fig = go.Figure(data = data, layout = layout)

pyo.iplot(fig)
#pyo.plot(data, filename ='scatter.html')
```

# Line charts

```python
import plotly
import plotly.offline as pyo
import plotly.graph_objs as go
import numpy as np

pyo.init_notebook_mode(connected=True)
```

```python
numberOfPoints = 200
values_x = np.linspace(0,1,numberOfPoints)
#print(values_x)
values_y = np.random.randn(numberOfPoints)

trace0 = go.Scatter(x= values_x, y=values_y + 5, 
                   mode='markers', name = 'markers')
trace1 = go.Scatter(x= values_x, y= values_y,
                   mode='lines', name = 'lines')
trace2 = go.Scatter(x = values_x, y= values_y -3, 
                   mode='lines+markers', name='lines and markers')
datalines = [trace0, trace1, trace2]
layoutLine= go.Layout(title='Line Charts')
fig = go.Figure(data=datalines, layout=layoutLine)

pyo.iplot(fig)
```

# Bart charts




```python
import plotly
import plotly.graph_objs as go
import plotly.offline as pyo
import pandas as pd
pyo.init_notebook_mode(connected=True)
```

```python
df_olympics = pd.read_csv('plotly/Data/2018WinterOlympics.csv')
df_olympics.head()
```

```python
data_olympics = [go.Bar(x=df_olympics['NOC'], y = df_olympics['Total'])]
layout_olympics = go.Layout(title='Olympics 2018 Medal count')
fig = go.Figure(data = data_olympics, layout = layout_olympics)
pyo.iplot(fig)
```

```python
trace_olympics_1 = go.Bar(x= df_olympics['NOC'], y = df_olympics['Gold'], name = 'Gold', marker ={'color': '#FFD700'})

trace_olympics_2 = go.Bar(x= df_olympics['NOC'], y = df_olympics['Silver'], name = 'Silver', marker ={'color': '#9EA0A1'})

trace_olympics_3 = go.Bar(x= df_olympics['NOC'], y = df_olympics['Bronze'], name = 'Bronze', marker ={'color': '#CD7F32'})


data_olympics2 = [trace_olympics_1, trace_olympics_2, trace_olympics_3]
layout_olympics2 = go.Layout(title='Olympics 2018 Medal count')
fig = go.Figure(data = data_olympics2, layout = layout_olympics2)
pyo.iplot(fig)
```

```python
layout_olympics3 = go.Layout(title='Olympics 2018 Medal count stacked', barmode='stack')
fig2 = go.Figure(data = data_olympics2, layout = layout_olympics3)
pyo.iplot(fig2)
```

# Bar chart exercise

```python
import plotly
import plotly.offline as pyo
import plotly.graph_objs as go
import pandas as pd
```

```python
df = pd.read_csv('plotly/Data/mocksurvey.csv', index_col=0)
#df = pd.read_csv('plotly/Data/mocksurvey.csv')
#df.set_index('Unnamed: 0', inplace=True)
df.head()

```

```python
data = [go.Bar(x=df.index, y = df[column], name = column) for column in df.columns]
layout = go.Layout(title='Survey', barmode='stack')
fig = go.Figure(data = data, layout= layout)
pyo.iplot(fig)
```

```python
data = [go.Bar(x=df[column], y = df.index, name = column, orientation='h') for column in df.columns]
layout = go.Layout(title='Survey', barmode='stack')
fig = go.Figure(data = data, layout= layout)
pyo.iplot(fig)
```
